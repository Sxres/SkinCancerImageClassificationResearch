```python
from ultralytics import YOLO
import os
import shutil
import random
```

```python
model = YOLO("yolo26m-cls.yaml")
```

```python
results = model.train(
    data='C:/Users/Dragos/OneDrive/Documents/SkinCancerImageClassificationResearch/DataCleaning/SkinCancer_YOLO_Split', 
    imgsz=448,           
    batch=16,             # 8 or 16 depending on gpu vram size
    workers=0,           
    device=0,            # forces gpu usage
    epochs=50,
    patience=10,
    lr0=0.0001,          
    optimizer='AdamW',
    
    augment=True,
    degrees=180.0,       # full rotation 
    fliplr=0.5,          # mirror left/right
    flipud=0.5,          # mirror up/down
    scale=0.6,           # zoom in/out heavily 
    hsv_h=0.015,         # slight color hue variance
    hsv_s=0.4,           # saturation variance (some images are redder)
    hsv_v=0.4,         
    
    project='skin_cancer_project',
    name='run_medium_448'      
)
```

```python
metrics = model.val(augment=True) 
print(f"Accuracy with TTA: {metrics.top1}")
```
